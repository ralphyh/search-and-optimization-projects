```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot for function X^2

```python
# Create a series from -20 to 20 and plot the function x**2
x = np.linspace(-10, 10, 40)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title('Graph of function x^2')
plt.grid()
plt.show()
```

## Plot for function X ^ 3

```python
# Create a series from -20 to 20 and plot the function x**3
x = np.linspace(-10, 10, 40)
y = x ** 3
```

```python
plt.plot(x, y)
plt.title('Graph of function x^3')
plt.grid()
plt.show()
```

## Plot for function x + x^4

```python
# Create a series from -20 to 20 and plot the function x + x^4
x = np.linspace(-10, 10, 40)
y = x + (x ** 4)
```

```python
plt.plot(x, y)
plt.title('Graph of function x + x^4')
plt.grid()
plt.show()
```

## Plot for function e^(x + x^2)

```python
# Create a series from -20 to 20 and plot the function e^(x + x^2)
x = np.linspace(-10, 10, 100)
y = np.exp(x + x**2)
```

```python
plt.plot(x, y)
plt.title('Graph of function e ^ (x + x^2)')
plt.grid()
plt.show()
```
