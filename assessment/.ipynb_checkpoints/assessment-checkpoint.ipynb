```python
import pandas as pd
import math
import folium
```

```python
base_station_1 = (50.817205047239135, -1.4493866785949845) # south-east end
base_station_2 = (50.86236620061749, -1.5720659114815352) # north-west end
central_coordinate = ((base_station_1[0] + base_station_2[0]) / 2, (base_station_1[1] + base_station_2[1]) / 2)
coords = pd.read_csv('sub_data_file.csv', names=["id", "x", "y"])
#map = folium.Map(location=central_coordinate, zoom_start=13, tiles="cartodb positron")
map = folium.Map(location=central_coordinate, zoom_start=13)
```

```python
def get_coordinates(coords):
    coordinates = []
    for i in coords.iterrows():
        coord = [i[1]["id"],(i[1]["y"]/110574 + central_coordinate[0], i[1]["x"]/111320 * math.cos(central_coordinate[0]) + central_coordinate[1])]
        coordinates.append(coord)

    return coordinates
```

```python
coordinates = get_coordinates(coords)
coordinates
```

```python
folium.Marker([50.817205047239135, -1.4493866785949845],  popup='Base Station 1').add_to(map)
folium.Marker([50.86236620061749, -1.5720659114815352],  popup='Base Station 2').add_to(map)
for location_info in coordinates:
    folium.CircleMarker([location_info[1][0], location_info[1][1]], radius=2, popup=location_info[0]).add_to(map)
```

```python
map.save("map.html")
```
