```python
def get_rank_weighting(population, selection_rate):
    n_keep = math.ceil(selection_rate * len(population))
    sorted_population = sorted(population,key=lambda rank:rank[2], reverse=False)[:n_keep]
    survivers = []
    rank_sum = 0
    for i in range(n_keep):
        rank_sum += i + 1
    for idx, i in enumerate(sorted_population):
        rank = idx + 1
        survivers.append(list([rank]))
        for j in i:
            survivers[idx].append(j)
        probability = round((n_keep  - rank + 1) / rank_sum, 4)
        survivers[idx].append(probability)

    return survivers
```

```python
def rastrigin_cost_func(population):
    population_cost = []
    for i in range(len(population)):
        population_cost.append(list([]))
        population_cost[i] = np.append(population_cost[i], round(population[i][0], 4))
        population_cost[i] = np.append(population_cost[i], round(population[i][1], 4))
        population_cost[i] = np.append(population_cost[i], round(20 + (population[i][0] ** 2 - \
                                                         (10 * math.cos(2 * math.pi * population[i][0]))) + \
                                                         (population[i][1] ** 2 - (10 * math.cos(2 * math.pi * population[i][1]))), 4))
    print("*----------------------------Performance------------------------------*")
    print("    Chromosomes           Scores")
    for i in population_cost:
        print("["+str(i[0])+", "+str(i[1])+"] ------> "+str(i[2]))
    print("*---------------------------------------------------------------------*")
    return population_cost
```

```python
def crossover_parent_pairing(selection_winners):
    np.random.shuffle(selection_winners)
    paired_parents = []
    for i in range(0, len(selection_winners) - len(selection_winners) % 2, 2):
        paired_parents.append([selection_winners[i], selection_winners[i + 1]])

    if len(selection_winners) % 2 != 0:
        paired_parents.append(selection_winners[-1])
    
    return paired_parents
```

```python
def crossover(paired_parents):
    new_generation = []
    for parents in paired_parents:
        try:
            rand = np.random.uniform(low=0.0, high=1.0, size=None)
            new_generation.append([round((rand * parents[0][0]) + (1 - rand) * parents[1][0], 4),\
             round((1 - rand) * parents[0][1] + (rand * parents[1][1]), 4)])
            new_generation.append(parents[0])
            new_generation.append(parents[1])
        except:
            new_generation.append(parents)

    return new_generation
```

```python
new_generation = crossover(paired_parents)
new_generation
```

```python
def mutation(population, mutation_sigma, lower_bound, upper_bound, mutation_rate=0.1):
    mutation_chromosome_indices = choice(np.array(range(len(population))),\
                                         math.ceil(len(population) * mutation_rate),\
                                         replace=False)
    for i in mutation_chromosome_indices:
        for idx, gene in enumerate(population[i]):
            mutation_point = randint(0, 2)
            if mutation_point == 1:
                valid_mutation = False
                while valid_mutation == False:       
                    rand = np.random.uniform(low=0.0, high=1.0, size=None)
                    new_gene = round(gene + (mutation_sigma * rand), 4)
                    if new_gene > lower_bound and new_gene < upper_bound:
                        population[i][idx] = new_gene
                        valid_mutation = True

    return population
```
